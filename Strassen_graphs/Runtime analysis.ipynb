```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
runtimes = pd.read_csv("report.csv")
```

```python
k = runtimes["K"].values
BAM = runtimes["BAM"].values
SAM = runtimes["SAM"].values
SAMK_K = runtimes["SAMK=K"].values
SAMK_K2 = runtimes["SAMK=K/2"].values
SAMK_K3 = runtimes["SAMK=sqrt(K)"].values
SAMK_K4 = runtimes["SAMK=LOG(K)"].values
```

```python


fig, ax = plt.subplots(figsize=(10,10))


ax.plot(k, BAM, label="BAM")
ax.plot(k, SAM, label="SAM")
ax.plot(k, SAMK_K, label="SAMK_K")
ax.plot(k, SAMK_K2, label="SAMK_k/2")
ax.plot(k, SAMK_K3, label="SAMK_sqrt(k)")
ax.plot(k, SAMK_K4, label="SAMK_log(K)")
ax.set_title('SAM vs BAM vs SAMKs')
ax.legend()
plt.grid(True)

plt.xlabel("K")
plt.ylabel("Time in seconds")

plt.show() 
fig.savefig("BAM_SAM_SAMK.jpg")
```

```python
def get_percent_diff(ori, opt):
    ans = []
    for i in range(len(ori)):
        ans.append((ori[i]-opt[i])/ori[i])
    return ans
```

```python
K_dif = get_percent_diff(BAM, SAMK_K)
K2_dif = get_percent_diff(BAM, SAMK_K2)
K3_dif = get_percent_diff(BAM, SAMK_K3)
K4_dif = get_percent_diff(BAM, SAMK_K4)
```

```python
print(K_dif)
print(K2_dif)
print(K3_dif)
print(K4_dif)
```
